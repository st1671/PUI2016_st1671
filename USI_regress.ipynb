```python
from pandas.stats.api import ols
from sklearn import linear_model
from sklearn import metrics

Number_variables=[]
Ridge_OS_F=[]
Ridge_IS_F=[]

Lasso_OS_F=[]
Lasso_IS_F=[]

OS_F=[]
IS_F=[]
t = 0
# Do a loop to run OLS after adding additional independent variables from 311 subsequently
for i in range(len(X_train.T)): #subsequently add variables
    
    t+=1
    Number_variables.append(t)
    
    X=X_train.iloc[:,:i+1]
    y=y_train

# In the sample:

    Ridge=linear_model.Ridge(fit_intercept=False,alpha=1000)
    Ridge.fit(X,y)
    
    p_IS=Ridge.predict(X)
    err_IS=p_IS-y
    R_2_IS=1-np.var(err_IS)/np.var(y)
    Ridge_IS_F.append(metrics.r2_score(y,p_IS))
    
    Lasso=linear_model.Lasso(fit_intercept=False, alpha=100)
    Lasso.fit(X,y)
    
    #In the sample:
    ls_IS=Lasso.predict(X)
    
    R_2_IS=metrics.r2_score(y,ls_IS)
    if R_2_IS>0:
        Lasso_IS_F.append(R_2_IS)
    else:
        Lasso_IS_F.append(0)

############################################################################
#Out of the sample test

    p_OS=Ridge.predict(X_test.iloc[:,:i+1])
    R_2_OS=metrics.r2_score(y_test,p_OS)
    
    if R_2_OS>0:
        Ridge_OS_F.append(R_2_OS)
    else:
        Ridge_OS_F.append(0)
        

    Lasso=linear_model.Lasso(fit_intercept=False, alpha=1000)
    Lasso.fit(X,y)
    
    ls_OS=Lasso.predict(X_test.iloc[:,:i+1])
    
    R_2_OS=metrics.r2_score(y_test,ls_OS)
    if R_2_OS>0:
        Lasso_OS_F.append(R_2_OS)
    else:
        Lasso_OS_F.append(1)


    result=ols(y=y_train,x=pd.DataFrame(X_train.iloc[:,:i+1]),intercept=False)
    R_2_IS=result.r2 
    
    a=np.array(X_test.iloc[:,:i+1])  
    b=np.array(result.beta) 
    c=np.sum(a*b,axis=1) #we do not add intercept here. c is predicted y. 
    error=y_test-c # y_test is real value, c is the value we predicted for y.
    R_2_OS=1-error.var()/y_test.var() # this is out of sample R2
    if R_2_OS>0:
        OS_F.append(R_2_OS)
    else:
        OS_F.append(0)

    IS_F.append(result.r2)
    
    
    

#plot
pylab.title('OS test for Ridge and Lasso ')

pylab.plot(Number_variables,IS_F,'g',label='R^2_OLS_IS')
pylab.plot(Number_variables,Lasso_IS_F,'b',label='R^2_Lasso_IS')
pylab.plot(Number_variables,Ridge_IS_F,'k',label='R^2_Ridge_IS')

pylab.plot(Number_variables,OS_F,'r',label='R^2_OLS_OS')
pylab.plot(Number_variables,Lasso_OS_F,'y',label='R^2_Lasso_OS')
pylab.plot(Number_variables,Ridge_OS_F,'c',label='R^2_Ridge_OS')

pylab.ylim(0, 0.7)


pylab.legend(loc='lower right')
pylab.xlabel('Number of independent variables')
pylab.ylabel('R-squared')
pylab.show()
```

```python
def Regularization_fit_lambda(model,X_train,y_train,lambdas,p=0.5,Graph=False, logl=False):
    #model = 1-Ridge, 2-Lasso
    #lambdas: a list of lambda values to try
    #p: ratio of the validation sample size / total training size
    #Graph: plot the graph of R^2 values for different lambda

    R_2_OS=[]
    X_train0, X_valid, y_train0, y_valid = train_test_split(X_train,
                                    y_train, test_size = 0.4, random_state = 200)

    if model==1:
        RM = lambda a: linear_model.Ridge(fit_intercept=True, alpha=a)
        model_label='Ridge'
    else:
        RM = lambda a: linear_model.Lasso(fit_intercept=True, alpha=a)
        model_label='Lasso'
    
    best_R2 = -1
    best_lambda = lambdas[0]
    
    for i in lambdas:
        lm = RM(i)
        lm.fit(X_train0,y_train0)  #fit the regularization model
        y_predict=lm.predict(X_valid) #compute the prediction for the validation sample 
        err_OS=y_predict-y_valid
        R_2_OS_=1-np.var(err_OS)/np.var(y_valid)
        R_2_OS.append(R_2_OS_)
        if R_2_OS_ > best_R2:
            best_R2 = R_2_OS_
            best_lambda = i
    
    if Graph==True:
        plt.title('IS R-squared vs OS-R-squared for different Lambda')
        if logl:
            plt.xlabel('ln(Lambda)')
            l=log(lambdas)
            bl=log(best_lambda)
        else:
            plt.xlabel('Lambda')
            l=lambdas
            bl=best_lambda
        plt.plot(l,R_2_OS,'b',label=model_label)
        plt.legend(loc='upper right')
        plt.ylabel('R-squared')
        plt.axvline(bl,color='r',linestyle='--')

        plt.show()
    
    return best_lambda
```
